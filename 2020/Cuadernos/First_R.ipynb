# Using IR (Kernel for R) with Colaboratory

Python is the kernel by default used by the **Jupyter** environment. However, we can also use another kernels (about a dozen of them).

One of them is known as the **IR** kernel.

Colab makes this process very easy (no installation required)



Code without visible output:

```r
a <- 8
```

```r
b <- 4:59
```

With visible output:

```r
head(a + b)
```

Printing is captured and sent to the frontend:

```r
print('Hello world! Love, R in Jupyter.')
```

So are errors:

```r
f2 <- function() stop('deep error')
throw <- function() f2()
```

```r
'this line is run / displayed'
throw()
'this line is not run / displayed'
```

```r
chol <- read.table(url("http://assets.datacamp.com/blog_assets/chol.txt"), header = TRUE, sep = " ")
head(chol)
hist(chol$AGE, main="Histogram of Age")
```

Plotting works too:

```r
x <- seq(0, 2*pi, length.out=50)
plot(x, sin(x))
```
